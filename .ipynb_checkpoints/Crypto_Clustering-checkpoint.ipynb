```python
# Import required libraries and dependencies
import os
#Set the environment variable to avoid the memory leak issue
os.environ['OMP_NUM_THREADS'] = '1'

import pandas as pd
import matplotlib.pyplot as plt
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
#df_market_data.hvplot.line(
    #width=800,
    #height=400,
    #rot=90 
#)

# Plot your data to see what's in your DataFrame
df_market_data.plot(kind='line', figsize=(10, 6))
plt.title('Cryptocurrency Market Data')
plt.xlabel('Coin ID')
plt.ylabel('Values')
plt.show()
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
# Fit and transform the data
crypto_scaled = StandardScaler().fit_transform(
    df_market_data[['price_change_percentage_24h',	'price_change_percentage_7d',	'price_change_percentage_14d',	
                    'price_change_percentage_30d',	'price_change_percentage_60d',	'price_change_percentage_200d',	'price_change_percentage_1y']]
)
```

```python
# Create a DataFrame with the scaled data
crypto_scaled_df = pd.DataFrame(
    crypto_scaled, 
    columns=['price_change_percentage_24h',	'price_change_percentage_7d',	'price_change_percentage_14d',	
                    'price_change_percentage_30d',	'price_change_percentage_60d',	'price_change_percentage_200d',	'price_change_percentage_1y']
)

# Copy the crypto names from the original data
crypto_scaled_df['coin_id'] = df_market_data.index

# Set the coinid column as index
crypto_scaled_df.set_index('coin_id', inplace=True)

# Display sample data
crypto_scaled_df.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 11))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
for k in k_values:
    
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
    kmeans = KMeans(n_clusters=k, random_state=0)
# 2. Fit the model to the data using `df_market_data_scaled`
    kmeans.fit(crypto_scaled_df)
# 3. Append the model.inertia_ to the inertia list
    inertia.append(kmeans.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k_values, "inertia": inertia}
# Create a DataFrame with the data to plot the Elbow curve
elbow_df = pd.DataFrame(elbow_data)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
plt.plot(elbow_df['k'], elbow_df['inertia'])
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.xticks(k_values)
plt.show()
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** The best value for 𝑘 is 4, as indicated by the Elbow method.

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state=0)
```

```python
# Fit the K-Means model using the scaled data
model.fit(crypto_scaled_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
crypto_clusters = model.predict(crypto_scaled_df)

# Print the resulting array of cluster values.
crypto_clusters
```

```python
# Create a copy of the DataFrame
crypto_df_copy = df_market_data.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
crypto_df_copy['Cluster'] = crypto_clusters

# Display sample data
crypto_df_copy.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
# Create the scatter plot using hvPlot
scatter_plot = crypto_df_copy.hvplot.scatter(
    x='price_change_percentage_24h',
    y='price_change_percentage_7d',
    by='Cluster',
    hover_cols=['crypto_name'],
    title='Cryptocurrency Clusters based on Price Changes',
    xlabel='24h Price Change (%)',
    ylabel='7d Price Change (%)'
)

# Display the plot
scatter_plot
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
crypto_pca = pca.fit_transform(crypto_scaled_df)

# Create a DataFrame with the PCA components
pca_df = pd.DataFrame(data=crypto_pca, columns=['PC1', 'PC2', 'PC3'])

# View the first five rows of the DataFrame. 
pca_df.head()
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
explained_variance = pca.explained_variance_ratio_

# View the explained variance
explained_variance
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** The total explained variance of the three principal components is approximately 0.895 (or 89.5%).

```python
# Create a new DataFrame with the PCA data.
pca_df = pd.DataFrame(data=crypto_pca, columns=['PC1', 'PC2', 'PC3'])

# Copy the crypto names from the original data
pca_df['coin_id'] = df_market_data.index

# Set the coinid column as index
pca_df.set_index('coin_id', inplace=True)

# Display sample data
pca_df.head()
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k_values_2 = list(range(1, 12))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
for k in k_values_2:
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
    kmeans = KMeans(n_clusters=k, random_state=0)
# 2. Fit the model to the data using `df_market_data_pca`
    kmeans.fit(crypto_pca)
# 3. Append the model.inertia_ to the inertia list
    inertia.append(kmeans.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {'k': k_values_2, 'inertia': inertia}
# Create a DataFrame with the data to plot the Elbow curve
elbow_df = pd.DataFrame(elbow_data)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
plt.figure(figsize=(10, 6))
plt.plot(elbow_df['k'], elbow_df['inertia'], marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.grid(True)
plt.show()
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** The best value for 𝑘 appears to be 4. This is the point where the inertia starts to decrease more slowly, indicating the optimal number of clusters.


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** The best value for 𝑘 found using the original data is also 4. Therefore, the best 𝑘 value does not differ between the PCA data and the original data. Both suggest that the optimal number of clusters is 4.

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
best_k = 4
kmeans_pca = KMeans(n_clusters=best_k, random_state=0)
```

```python
# Fit the K-Means model using the PCA data
kmeans_pca.fit(crypto_pca)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
pca_clusters = kmeans_pca.predict(crypto_pca)

# Print the resulting array of cluster values.
pca_clusters
```

```python
# Create a copy of the DataFrame with the PCA data
pca_df_copy = pca_df.copy()

# Add a new column to the DataFrame with the predicted clusters
pca_df_copy['Cluster'] = pca_clusters

# Display sample data
pca_df_copy.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
scatter_plot = pca_df_copy.hvplot.scatter(
    x='PC1',
    y='PC2',
    by='Cluster',
    hover_cols=['coin_id'],
    title='Cryptocurrency Clusters based on PCA Data',
    xlabel='Principal Component 1 (PC1)',
    ylabel='Principal Component 2 (PC2)'
)
# Display the plot
scatter_plot
```

### Visualize and Compare the Results

In this section, you will visually analyze the cluster analysis results by contrasting the outcome with and without using the optimization techniques.

```python
# Composite plot to contrast the Elbow curves
# Scale the original data
scaler = StandardScaler()
crypto_scaled_df = scaler.fit_transform(
    df_market_data[['price_change_percentage_24h', 'price_change_percentage_7d', 
                    'price_change_percentage_14d', 'price_change_percentage_30d', 
                    'price_change_percentage_60d', 'price_change_percentage_200d', 
                    'price_change_percentage_1y']]
)

# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))

# Compute inertia values for the original data
inertia_original = []
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(crypto_scaled_df)
    inertia_original.append(kmeans.inertia_)

# Create a DataFrame with the inertia values for original data
df_elbow = pd.DataFrame({'k': k_values, 'inertia': inertia_original})

# Apply PCA to reduce dimensions
pca = PCA(n_components=3)
crypto_pca = pca.fit_transform(crypto_scaled_df)

# Compute inertia values for the PCA data
inertia_pca = []
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(crypto_pca)
    inertia_pca.append(kmeans.inertia_)

# Create a DataFrame with the inertia values for PCA data
pca_df_elbow = pd.DataFrame({'k': k_values, 'inertia': inertia_pca})

# Create a DataFrame with the inertia values for original data
df_elbow = pd.DataFrame({'k': k_values, 'inertia': inertia_original})

plots = (df_elbow.hvplot.line(x="k", y="inertia", title="Original Elbow Curve") + 
         pca_df_elbow.hvplot.line(x="k", y="inertia", title="PCA Elbow Curve")).cols(1)
plots
```

```python
# Composite plot to contrast the clusters

# Plot for original data clusters
original_clusters_plot = crypto_df_copy.hvplot.scatter(
    x='price_change_percentage_24h',
    y='price_change_percentage_7d',
    by='Cluster',
    hover_cols=['coin_id'],
    title='Cryptocurrency Clusters based on Original Data',
    xlabel='24h Price Change (%)',
    ylabel='7d Price Change (%)'
)

# Create a copy of the PCA DataFrame with cluster labels
pca_df_copy = pd.DataFrame(crypto_pca, columns=['PC1', 'PC2', 'PC3'])
pca_df_copy['coin_id'] = df_market_data.index
pca_df_copy.set_index('coin_id', inplace=True)
pca_df_copy['Cluster'] = KMeans(n_clusters=4, random_state=0).fit_predict(crypto_pca)

# Plot for PCA data clusters
pca_clusters_plot = pca_df_copy.hvplot.scatter(
    x='PC1',
    y='PC2',
    by='Cluster',
    hover_cols=['coin_id'],
    title='Cryptocurrency Clusters based on PCA Data',
    xlabel='Principal Component 1 (PC1)',
    ylabel='Principal Component 2 (PC2)'
)

# Combine the plots into a composite plot
composite_plot = original_clusters_plot + pca_clusters_plot

# Display the composite plot
composite_plot
```

#### Answer the following question: 

  * **Question:** After visually analyzing the cluster analysis results, what is the impact of using fewer features to cluster the data using K-Means?

  * **Answer:** Using fewer features to cluster the data with K-Means has several noticeable impacts:

Cluster Separation: With PCA, which reduces the data to three principal components, the clusters appear more separated and distinct. This indicates that the PCA transformation helps to better differentiate between the clusters by capturing the most important variance in the data.

Compact Clusters: The clusters formed using PCA are more compact, suggesting that the key features extracted by PCA contribute significantly to forming well-defined clusters. This compactness may lead to more reliable and interpretable clustering results.

Feature Influence: The original data, which includes all features without dimensionality reduction, shows more dispersed clusters. This dispersion can be attributed to the presence of noise and less relevant features, which may obscure the clustering structure.

Elbow Curve Analysis: The Elbow curves for both the original and PCA-transformed data indicate that the optimal number of clusters is 4. However, the PCA data's Elbow curve shows a steeper initial drop, suggesting that the principal components capture the variance more effectively and make it easier to identify the optimal number of clusters.

In summary, using PCA to reduce the number of features helps to form more distinct and compact clusters, leading to potentially more accurate and interpretable clustering results with K-Means.
